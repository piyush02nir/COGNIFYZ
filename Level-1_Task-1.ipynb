### Replace missing values with the mode.

```python
cuisine_mode=data.Cuisines.mode()[0]
cuisine_mode
```

```python
data.Cuisines.fillna(cuisine_mode,inplace=True)
data.Cuisines.isnull().sum()
```

```python
data.isnull().sum()
```

```python
# Set a custom style for the plot (optional)
sns.set_style("whitegrid")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create a histogram with specified bins and edgecolor
sns.histplot(data=data, x='Aggregate rating', bins=10, edgecolor='k', ax=ax)

# Set axis labels and title
ax.set_xlabel('Aggregate Rating', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
ax.set_title('Distribution of Aggregate Ratings', fontsize=16)

# Customize tick label font size (optional)
ax.tick_params(axis='both', labelsize=12)

# Customize grid lines (optional)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
